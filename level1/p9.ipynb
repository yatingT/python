```python
def clean(data):
    clean=[]
    for row in data:
        try:
            for i in row:
                assert row[i] != None
            clean.append(row)
        except:
            continue
    return clean

def get_reviews(csv_path,json_path):
    try:
        json=read_json(os.path.join("data",json_path))
    except:
        json=json_path
    temp=read_csv("data",csv_path)
    csv=clean(temp)
    final=[]
    for i in csv:
        if i['review do_recommend']=='False':
            i['review do_recommend'] = ""
        if json == None:
            continue
        try:
            g=Review(id=int(i['review id']),
                 username=json[i['review id']][0],
                 asin=json[i['review id']][1],
                title=i['review title'],
                 text=i['review text'],
                 rating=int(i['review rating']),
                do_recommend=bool(i['review do_recommend']),
                 num_helpful=int(i['review num_helpful']),
                date=i['review date'])
            final.append(g)
        except:
            continue
    return final
#clean(read_csv("data",'review2.csv'))
get_reviews('review2.csv', 'review2.json')[-10:] 
```

```python
import pandas as pd

def scatter(x, y, xlabel="please label me!", ylabel="please label me!"):
    df = pd.DataFrame({"x":x, "y":y})
    ax = df.plot.scatter(x="x", y="y", color="black", fontsize=16, xlim=0, ylim=0)
    ax.set_xlabel(xlabel, fontsize=16)
    ax.set_ylabel(ylabel, fontsize=16)
    ax.get_xaxis().get_major_formatter().set_scientific(False)
    ax.get_yaxis().get_major_formatter().set_scientific(False)
```

```python
#q27
def bucketize(category):
    bucket={}
    key=list(set(df[category]))
    for rate in key:
        temp=[]
        for idx in range(len(df)):
            if df.loc[idx][category]==rate:
                temp.append(df.loc[idx])
        bucket[str(rate)]=temp
    return bucket


def get_an(category,in_list):
    an={}
    x=bucketize(category)
    for bucket in x:
        total=0
        for row in x[bucket]:
            total+=len(row[in_list])
            average=total/len(x[bucket])
        an[str(bucket)]=average  
    return an

plot=get_an("rating","text")
key=[]
value=[]
for i in plot:
    key.append(int(i))
    value.append(int(plot[i]))

scatter(key,
        value,
       xlabel="rating", 
        ylabel="average taxt length")

```

```python
def file_path(path):
    bf=os.listdir(path)
    real_path=[]
    for i in bf:
        if os.path.isfile(os.path.join(path,i)):
            real_path+=[os.path.join(path,i)]
        else:
            real_path+=file_path(os.path.join(path,i))
    return real_path
```
